# Clustering

### Is Done Using K Means Algorithm

###### Python Solution for Clustering in Unsupervised Learning

```python
# Importing the libraries

import pandas as pd
```

### K Means Library Below

```python
# Importing the K means library

from sklearn.cluster   import KMeans

from sklearn.preprocessing import StandardScaler
```

### Importing the file

```python
# File Importing is done this way

dfile = pd.read_csv("Mall_Customers.csv")
dfile.head()
```

### Renaming the columns

```python
dfile.columns
```

```python
dfile = dfile.rename(columns={
    'Annual Income (k$)':  'Annual_Income',
    'Spending Score (1-100)': 'Spend_Score'
})
```

### Splitting the features

```python
x = dfile[['Age', 'Annual_Income', 'Spend_Score']]

x.head()
```

##### Since Unsupervised Learning doesn't have output variables to train the model and make prediction we use the concept to understand our datasets.

### Scaling the x values

```python
scaler = StandardScaler()
xscaled = scaler.fit_transform(x)
xscaled


xscaled.dtype

```

### Converting the numpy array to dataframe

```python
x_new = pd.DataFrame(xscaled, columns= ['x', 'y', 'z'])
x_new
```

### Fitting the model

```python
model = KMeans(n_clusters = 5)
model.fit(x_new)
```

### Prediction is Not Done But Clustering is Done in Here

##### Labels_ is Used For Clustering

```python
labels = model.labels_
```

### Data Visualization

```python
# Importing the matplotlbpyplot libb 

import matplotlib.pyplot as plt

plt.scatter(x_new.x, x_new.y, x_new.z, c= labels)\
```
